```python
import time
import random

# Helper function to print and pause
def print_pause(message, delay=2):
    print(message)
    time.sleep(delay)

# Global variables for game settings
MAX_TURNS = 5
WINNING_SCORE = 5

# Scenario One: The stormy bridge
def stormy_bridge(score):
    print_pause("\nYou approach a creaky bridge under a raging storm.")
    danger = random.choice(["strong wind", "cracking wood", "a sudden lightning strike"])
    print_pause(f"As you think, you notice: {danger}")

    while True:
        choice = input("Do you (1) Cross the bridge or (2) Wait for the storm to pass? (Enter 1 or 2): ")
        if choice == "1":
            if danger == "cracking wood":
                print_pause("The bridge breaks! You fall and get hurt.")
                score -= 1
            else:
                print_pause("You make it across safely and feel braver.")
                score += 1
            break
        elif choice == "2":
            print_pause("You wait patiently. The storm eases a little.")
            score += 2
            break
        else:
            print("Please enter 1 or 2.")
    return score

# Scenario Two: glowing cave
def glowing_cave(score):
    print_pause("\nYou step into a quiet forest. The trees seem to whisper secrets.")
    whisper = random.choice(["a warning", "a riddle", "a curse"])
    print_pause(f"You hear: '{whisper}' carried in the wind.")

    while True:
        choice = input("Do you (1) Listen closely or (2) Ignore the whisper? (Enter 1 or 2): ")
        if choice == "1":
            if whisper == "a curse":
                print_pause("You feel a chill, something drains your energy.")
                score -= 2
            else:
                print_pause("You feel enlightened by what you hear.")
                score += 2
            break
        elif choice == "2":
            print_pause("You move on. It may have been wise.")
            score += 1
            break
        else:
            print("Please enter 1 or 2.")
    return score

# Main game loop
def start_game():
    while True:
        total_score = 3
        turns = 0

        print_pause("\n You find yourself in the wild of Fate!")
        print_pause("To win, you must survive and obtain a score of 5.")

        while turns < MAX_TURNS:
            print_pause(f"\nTurn {turns + 1}. Your score is now: {total_score}")
            choice = input("Do you (1) Cross the stormy bridge or (2) Enter the whispering grove? (Enter 1 or 2): ")

            if choice == "1":
                total_score = stormy_bridge(total_score)
                turns += 1
            elif choice == "2":
                total_score = glowing_cave(total_score)
                turns += 1
            else:
                print("Please enter 1 or 2.")
                continue

            if total_score >= WINNING_SCORE:
                print_pause("\n Congratulations! You have survived and won the game!")
                print_pause("\n Your Total Score is " + str(total_score))
                break
            elif total_score <= 0:
                print_pause("\n All your strength has faded. Game over.")
                break

        if total_score > 0 and total_score < WINNING_SCORE and turns == MAX_TURNS:
            print_pause("\n Time's up! You didn't gather enough points. Game over.")

        # Replay option
        while True:
            again = input("\nWould you like to play again? (y/n): ").lower()
            if again == 'y':
                break  # Will restart the game loop
            elif again == 'n':
                print_pause("Thanks for playing! Goodbye.")
                return  # Exits the entire function
            else:
                print("Please enter 'y' or 'n'.")

# Start the game
start_game()
```
